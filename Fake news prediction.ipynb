```python
import pandas as pd
import numpy as np
```

```python
newsdata=pd.read_csv("news.csv")
newsdata
```

```python
newsdata.isnull().sum()
```

```python
newsdata['label'] = newsdata['label'].astype('category')
newsdata['label'] = newsdata['label'].cat.codes
newsdata
```

```python
inputx=newsdata.drop(columns=["Unnamed: 0"],axis="columns")
inputx
```

```python
outputy=newsdata["label"]
outputy
```

```python
newsdata.describe().astype(int)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(newsdata["text"],outputy,test_size=0.25,random_state=0)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer=TfidfVectorizer(stop_words='english', max_df=0.7)


tfidf_train=tfidf_vectorizer.fit_transform(x_train) 
tfidf_test=tfidf_vectorizer.transform(x_test)
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
pac=PassiveAggressiveClassifier(max_iter=50)
pac.fit(tfidf_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
y_pred=pac.predict(tfidf_test)
score=accuracy_score(y_test,y_pred)
print(f'Accuracy: {(score*100,2)}%')
```
